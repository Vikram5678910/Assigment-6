```python
import pandas as pd
import numpy as np
import torch
from google.colab import files
from datasets import Dataset
from sklearn.model_selection import train_test_split
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
import evaluate
from transformers import TrainingArguments, Trainer
```

```python
uploaded = files.upload()
filename = list(uploaded.keys())[0]
df = pd.read_csv(filename)
```

```python
print("Shape:", df.shape)
print("\nColumns:", df.columns.tolist())
display(df.head())

print("\nNull counts per column:\n", df.isnull().sum())
print("\nNumber of duplicate rows:", df.duplicated().sum())

print("\nSample rows:")
display(df.sample(5, random_state=42))
```

```python
df = df.drop_duplicates()
df = df.dropna(subset=['question', 'answer'])
```

```python
def clean_text(text):
    text = str(text).strip()
    text = text.replace("\n", " ")
    text = text.replace("\r", " ")
    text = text.replace("\t", " ")
    return text

df['question'] = df['question'].apply(clean_text)
df['answer'] = df['answer'].apply(clean_text)

print("\nAfter cleaning:")
display(df.head(3))
print("Shape:", df.shape)
```

```python
from transformers import AutoTokenizer
tokenizer = AutoTokenizer.from_pretrained("google/flan-t5-small")

def count_tokens(row):
    text = f"Question: {row['question']} Answer: {row['answer']}"
    return len(tokenizer.encode(text))

df['token_count'] = df.apply(count_tokens, axis=1)

print("Token count summary:")
print(df['token_count'].describe())
print("Total tokens in dataset:", df['token_count'].sum())
```

```python
train_df, temp_df = train_test_split(df, test_size=0.2, random_state=42)
val_df, test_df = train_test_split(temp_df, test_size=0.5, random_state=42)

print("Train size:", len(train_df))
print("Validation size:", len(val_df))
print("Test size:", len(test_df))
```

```python
df["input_text"] = df.apply(lambda x: f"Question about {x['focus_area']}: {x['question']}", axis=1)
df["target_text"] = df["answer"]

train_df, temp_df = train_test_split(df, test_size=0.2, random_state=42)
val_df, test_df = train_test_split(temp_df, test_size=0.5, random_state=42)

train_dataset = Dataset.from_pandas(train_df[['input_text', 'target_text']])
val_dataset = Dataset.from_pandas(val_df[['input_text', 'target_text']])

print("Train samples:", len(train_dataset))
print("Validation samples:", len(val_dataset))
```

```python
model_name = "google/flan-t5-small"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

print("Using device:", device)
```

```python
def preprocess_function(batch):
    inputs = tokenizer(
        batch["input_text"],
        padding="max_length",
        truncation=True,
        max_length=128
    )
    targets = tokenizer(
        batch["target_text"],
        padding="max_length",
        truncation=True,
        max_length=128
    )
    inputs["labels"] = targets["input_ids"]
    return inputs
```

```python
tokenized_train = train_dataset.map(preprocess_function, batched=True)
tokenized_val = val_dataset.map(preprocess_function, batched=True)

print("Tokenization complete!")
print("Train samples:", len(tokenized_train))
print("Validation samples:", len(tokenized_val))
```

```python
training_args = TrainingArguments(
    output_dir="/content/flan_t5_small_results",
    num_train_epochs=2,
    per_device_train_batch_size=1,
    per_device_eval_batch_size=1,
    logging_steps=100,
    save_steps=500,
    learning_rate=2e-4,
    do_train=True,
    do_eval=True,
    report_to=[]
)

```

```python
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_train,
    eval_dataset=tokenized_val,
    tokenizer=tokenizer,
)
```

```python
rouge = evaluate.load("rouge")

def compute_metrics(eval_pred):
    preds, labels = eval_pred
    preds = tokenizer.batch_decode(preds, skip_special_tokens=True)

    labels = [[l for l in label if l != -100] for label in labels]
    labels = tokenizer.batch_decode(labels, skip_special_tokens=True)

    return rouge.compute(predictions=preds, references=labels)
```

```python
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_train,
    eval_dataset=tokenized_val,
    tokenizer=tokenizer,
    compute_metrics=compute_metrics,
)

print("Trainer ready!")
```

```python
%%time
result = trainer.train()
trainer.save_model("/content/flan_t5_small_model")

print("✅ MODEL TRAINING COMPLETE & SAVED!")
```

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
import torch

# Load your fine-tuned model
model_path = "/content/flan_t5_small_model"

tokenizer = AutoTokenizer.from_pretrained(model_path)
model = AutoModelForSeq2SeqLM.from_pretrained(model_path)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

print("Model loaded on:", device)
```

```python
def answer_medical_question(question, focus_area="General Health"):
    input_text = f"Question about {focus_area}: {question}"

    inputs = tokenizer(input_text, return_tensors="pt").to(device)

    outputs = model.generate(
        **inputs,
        max_length=256,
        num_beams=5,              # better exploration
        repetition_penalty=2.5,   # stops repeating words
        no_repeat_ngram_size=3,   # avoids repeated phrases
        length_penalty=1.0,
        early_stopping=True
    )

    return tokenizer.decode(outputs[0], skip_special_tokens=True)
```

```python
print(answer_medical_question("What are the symptoms of diabetes?", "Diabetes"))
```

```python
from datasets import Dataset
import numpy as np

def compute_rouge_on_test():
    model.eval()

    preds = []
    refs = []

    for i in range(200):   # evaluate on first 200 samples for speed
        q = test_df.iloc[i]["question"]
        focus = test_df.iloc[i]["focus_area"]
        gold = test_df.iloc[i]["answer"]

        pred = answer_medical_question(q, focus)

        preds.append(pred)
        refs.append(gold)

    results = rouge.compute(predictions=preds, references=refs)
    return results

rouge_results = compute_rouge_on_test()
print(rouge_results)
```

```python
from deep_translator import GoogleTranslator

def translate_to_english(text):
    return GoogleTranslator(source='auto', target='en').translate(text)

def translate_from_english(text, target_lang):
    return GoogleTranslator(source='en', target=target_lang).translate(text)
```

```python
from langdetect import detect

def detect_language(text):
    try:
        return detect(text)
    except:
        return "en"
```

```python
def multilingual_medical_chatbot(user_input):
    # 1. Detect language
    lang = detect_language(user_input)

    # 2. Translate to English if needed
    if lang != "en":
        english_input = translate_to_english(user_input)
    else:
        english_input = user_input

    # 3. Generate medical answer (your model)
    answer_en = answer_medical_question(english_input, "")

    # 4. Translate back to original language
    if lang != "en":
        final_answer = translate_from_english(answer_en, lang)
    else:
        final_answer = answer_en

    return final_answer
```
